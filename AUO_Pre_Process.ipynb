```python
#離散型資料補眾數
df['Cat_01'] = df['Cat_01'].fillna(df['Cat_01'].mode()[0])
df['Cat_02'] = df['Cat_02'].fillna(df['Cat_02'].mode()[0])
df['Cat_03'] = df['Cat_03'].fillna(df['Cat_03'].mode()[0])
df['Cat_04'] = df['Cat_04'].fillna(df['Cat_04'].mode()[0])
df['Cat_05'] = df['Cat_05'].fillna(df['Cat_05'].mode()[0])
df['Cat_06'] = df['Cat_06'].fillna(df['Cat_06'].mode()[0])
df['Cat_07'] = df['Cat_07'].fillna(df['Cat_07'].mode()[0])
df['Cat_08'] = df['Cat_08'].fillna(df['Cat_08'].mode()[0])
df['Cat_09'] = df['Cat_09'].fillna(df['Cat_09'].mode()[0])
df['Cat_10'] = df['Cat_10'].fillna(df['Cat_10'].mode()[0])
```

```python
#獲取column name
column_name = []
for col in df.columns:
    column_name.append(col)
print(len(column_name))
```

```python
#數值型資料column name
col_column_name = column_name[12:]
```

```python
#補中位數與平均數的column name
fill_na_median = []
fill_na_mean = []
for i in col_column_name:
    if abs(df["{}".format(i)].skew()) > 2:
        fill_na_median.append(i)
    else:
        fill_na_mean.append(i)
print(len(fill_na_mean))
print(len(fill_na_median))
```

```python
#平均數補值
for i in fill_na_mean:
    a = df["{}".format(i)].mean()
    df["{}".format(i)].fillna(a, inplace = True)
    
#中位數補值
for i in fill_na_median:
    a = df["{}".format(i)].median()
    df["{}".format(i)].fillna(a, inplace = True)
```

```python
#離散資料編碼
dum = pd.get_dummies(df, columns=['Cat_01', 'Cat_02', 'Cat_03', 'Cat_04', 'Cat_05', 'Cat_06', 'Cat_07', 'Cat_08', 'Cat_09', 'Cat_10'])
pd.DataFrame(dum)
```

```python
#標準化
for i in col_column_name:
    dum["{}".format(i)] = (dum["{}".format(i)] - dum["{}".format(i)].mean()) / (df["{}".format(i)].std())
dum['Col_0001'].std()  
```
